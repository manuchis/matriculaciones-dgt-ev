```python
def import_file(file):
    print('Adding file',file)
    df = pd.read_fwf(dire+file,header=None,colspecs=colspecs, names=cols, encoding='Latin-1')
    df = df[~df.FEC_MATRICULA.str.contains('|'.join(['Vehícul']))] #borra la linea de texto inicial
    #df['FEC_MATRICULA'] = pd.to_datetime(df['FEC_MATRICULA'], format='%d%m%Y')
    #df['FEC_TRAMITACION'] = pd.to_datetime(df['FEC_TRAMITACION'], format='%d%m%Y')
    #df['FEC_PRIM_MATRICULACION'] = pd.to_datetime(df['FEC_PRIM_MATRICULACION'], format='%d%m%Y')
    #df['FEC_PROCESO'] = pd.to_datetime(df['FEC_PROCESO'], format='%d%m%Y')
    return df
    
```

```python
dfg = pd.DataFrame(columns=cols)
for f in files: 
    df= import_file(f)
    dfg = pd.concat([dfg, df])
    print('File added to list')

dfg.info()
```

```python
# exportar registros enteros
dfg.to_json('registros.json', orient='records')
```
